# Data Exploration!

```python
import pandas
import sklearn
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
import os
import matplotlib.pyplot as plt
```

```python
def get_tag_counts_for_text(story_string):
    text = nltk.word_tokenize(story_string)
    word_tag_tuples = nltk.pos_tag(text)
    all_tags = [tag for _, tag in word_tag_tuples]
    possible_tags = set(all_tags)
    tag_counts = dict()
    for tag in all_tags:
        if tag in tag_counts.keys():
            tag_counts[tag] += 1
        else:
            tag_counts[tag] = 0
    return tag_counts

flatten = lambda l: [item for sublist in l for item in sublist]
    
```

```python
qualities = ['good', 'medium', 'bad']    
stories = flatten([[{"quality": quality, "filename": file} for file in os.listdir(quality)] for quality in qualities])


for story in stories:
        with open(story["quality"] + '/' + story["filename"]) as f:
            story["tag_counts"] = get_tag_counts_for_text(f.read())
```

```python
for story in stories:
    story['normalized_counts'] = {tag: count/sum(story['tag_counts'].values()) for tag, count in story['tag_counts'].items()}
```

```python
all_tags = []

for story in stories:
    all_tags.append(list(story['normalized_counts'].keys()))
    
all_tags = list(set(flatten(all_tags)))
all_tags.sort()
corr = {tag: count for count, tag in enumerate(all_tags)}
corr
```

```python
for story in stories:
    for tag, item in story['normalized_counts'].items():
        print(corr[tag])
```

```python
for story in stories:
    story['enumerated_tag_counts'] = {corr[tag]:normalized_count for tag, normalized_count in story['normalized_counts'].items()}
```

```python
for story in stories:
    for i in corr.values():
        if not i in story['enumerated_tag_counts'].keys():
            story['enumerated_tag_counts'][i] = 0
```

```python
for story in stories:
    to_plot = story['enumerated_tag_counts']
    x = sorted(to_plot.keys())
    y = [to_plot[i] for i in x]
    print(story['quality'])
    print(story['filename'])
    plt.bar(x, y)
    plt.show()
```

```python
flattened_stories = [
    [story['filename'], story['quality']] + [
        story['enumerated_tag_counts'][key] for key in sorted(story['enumerated_tag_counts'].keys())
    ]
    for story in stories
]

columns = ['filename', 'quality'] + list(stories[0]['enumerated_tag_counts'].keys())
pandas.DataFrame(flattened_stories, columns=columns)
```
