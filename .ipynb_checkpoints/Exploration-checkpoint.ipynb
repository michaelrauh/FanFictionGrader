# Data Exploration!

```python
import sklearn
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
import os
```

```python
def get_tag_counts_for_text(story_string):
    text = nltk.word_tokenize(story_string)
    word_tag_tuples = nltk.pos_tag(text)
    all_tags = [tag for _, tag in word_tag_tuples]
    possible_tags = set(all_tags)
    tag_counts = dict()
    for tag in all_tags:
        if tag in tag_counts.keys():
            tag_counts[tag] += 1
        else:
            tag_counts[tag] = 0
    return tag_counts
```

```python
qualities = ['good', 'medium', 'bad']    
quality_to_filenames = {quality: os.listdir(quality) for quality in qualities}
quality_to_stories = {quality: [] for quality in qualities}


for quality, filenames in quality_to_filenames.items():
    for filename in filenames:
        with open(quality + '/' + filename) as f:
            quality_to_stories[quality].append(f.read())
```

```python
quality_to_tag_counts = {quality: [] for quality in qualities}

for quality, stories in quality_to_stories.items():
    for story in stories:
        quality_to_tag_counts[quality].append(get_tag_counts_for_text(story))
```
